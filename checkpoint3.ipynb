```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
import seaborn as sns
house=pd.read_csv('kc_house_data.csv')
house.head()
```

```python
house.info()
house.describe()
```

```python
house.isnull().sum()
```

```python
house=house[['price', 'bedrooms', 'bathrooms', 'sqft_living',
    'sqft_lot', 'floors', 'waterfront', 'view', 'condition', 'grade',
    'sqft_above', 'sqft_basement', 'yr_built', 'yr_renovated', 'zipcode',
    'lat', 'long', 'sqft_living15', 'sqft_lot15']]
h = house.hist(bins=25,figsize=(16,16),xlabelsize='10',ylabelsize='10',xrot=-15)
sns.despine(left=True, bottom=True)
[x.title.set_size(12) for x in h.ravel()];
[x.yaxis.tick_left() for x in h.ravel()];
```

```python
house_df=house.copy()
```

```python
def plot_correlation_map( df ):

    corr = df.corr()

    s , ax = plt.subplots( figsize =( 17 , 15 ) )

    cmap = sns.diverging_palette( 220 , 10 , as_cmap = True )

    s = sns.heatmap(

        corr, 

        cmap = cmap,

        square=True, 

        cbar_kws={ 'shrink' : .9 }, 

        ax=ax, 

        annot = True, 

        annot_kws = { 'fontsize' : 12 }

        )

plot_correlation_map(house_df)
```

According to the heatmap, we can notice that there is a strong correlation between the sqft_living15, the sqft_living,the sqft above  the grade  and  the price. We also notice that there is a moderate correlation between the view and the price, the bathroom and the price but we will take into consideration as we see this feature plays a role in determining the price of the house


```python
house_df=house_df.drop(columns=["sqft_lot","floors","waterfront","condition","sqft_basement","yr_built","yr_renovated","lat","long","sqft_lot15","bedrooms"],axis=1)

```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn import metrics
#extract x and y from our data
x=house_df["sqft_living"].values[:,np.newaxis]
y=house_df["price"].values

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=80) #splitting data with test size of 25%
model=LinearRegression()   #build linear regression model
model.fit(x_train,y_train)  #fitting the training data
predicted=model.predict(x_test) #testing our modelâ€™s performance



```

```python
plt.scatter(x,y,color="r")
plt.title("Linear Regression")
plt.ylabel("price")
plt.xlabel("sqft_living")
plt.plot(x,model.predict(x),color="k")
plt.show()
```

```python
print("MSE", mean_squared_error(y_test,predicted))
print("R squared", metrics.r2_score(y_test,predicted))
```

We can conclude that the linear regression model is not the best model to use

```python
x=house_df[["grade","sqft_living15", "sqft_living","sqft_above","view","bathrooms"]]  #we have more than one input
y=house_df["price"].values
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=40) #splitting data with test size of 25%

model=LinearRegression() #build linear regression model
model.fit(x_train,y_train) #fitting the training data
predicted=model.predict(x_test) #testing our modelâ€™s performance

print("MSE", mean_squared_error(y_test,predicted))
print("R squared", metrics.r2_score(y_test,predicted))
```

The MSE is a large number we can also conclude that the multiple linear regression is not the appropriate model to use

```python
from sklearn.preprocessing import PolynomialFeatures 
x=house_df[["grade","sqft_living15", "sqft_living","sqft_above","bathrooms","view"]]  #we have more than one input

y=house_df["price"].values
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=40) #splitting data with test size of 25%

lg=LinearRegression()
poly=PolynomialFeatures(degree=2)

x_train_fit = poly.fit_transform(x_train) #transforming our input data
lg.fit(x_train_fit, y_train)
x_test_ = poly.fit_transform(x_test)
predicted = lg.predict(x_test_)

print("MSE: ", metrics.mean_squared_error(y_test, predicted))
print("R squared: ", metrics.r2_score(y_test,predicted))
```

```python
x= house_df['sqft_living'].values.reshape(-1,1)
y= house_df["price"].values
poly = PolynomialFeatures(degree = 2) 
x_poly = poly.fit_transform(x) 
poly.fit(x_poly, y) 
lg=LinearRegression()
lg.fit(x_poly, y) 

plt.scatter(x, y, color="r")
plt.title("Linear regression")
plt.ylabel("price")
plt.xlabel("sqft_living")
plt.plot(x, lg.predict(poly.fit_transform(x)), color="k") 
```

the R squared of the polynominial regression is the closest to one compared to the linear and multiple linear regression models so far. We can conclude that it is the best model we can work with
